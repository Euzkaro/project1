```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from geopy.distance import vincenty
```

```python
## Import Data
# -----------------------------
# 1) Yelp Data Chicago
Yelp_df = pd.read_csv("Yelp_Restaurants_Chicago.csv")
# 2) Business License Chicago
License_df = pd.read_csv("Clean_Chicago_Restaurants_Licenses.csv")
len(License_df)
```

```python
License_Clean_df = License_df.groupby("DOING BUSINESS AS NAME").first().reset_index()
len(License_Clean_df)
```

```python
print(License_Clean_df.columns)
print(Yelp_df.columns)
```

```python
# Initialize the DataFrame
Yelp_License_df = Yelp_df
Yelp_License_df["Min_Dist"] = np.nan
Yelp_License_df["Min_Name"] = np.nan

Yelp_License_df.head()
```

```python
for index, row in Yelp_License_df.iterrows():
    lat = row["latitude"]
    lng = row["longitude"]
    name = row["name"]
    dist_list = License_Clean_df.apply(lambda x: vincenty((x["LATITUDE"], x["LONGITUDE"]), (lat,lng)).meters, axis = 1)
    dist_min = min(dist_list)
    dist_min_index = dist_list[dist_list==dist_min].index[0]
    dist_min_name = License_Clean_df.loc[dist_min_index,"DOING BUSINESS AS NAME"]
    Yelp_License_df.loc[index,"Min_Dist"] = dist_min
    Yelp_License_df.loc[index,"Min_Name"] = dist_min_name
    Yelp_License_df.loc[index,"Date_Issued"] = License_Clean_df.loc[dist_min_index,"DATE ISSUED"]
    print(f"Restaurant name is {name}, the closest business is {dist_min_name} with dist as {dist_min}")
    
Yelp_License_df.to_csv("Yelp_License_Merge.csv",index=False)
```
